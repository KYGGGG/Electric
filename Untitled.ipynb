# 데이터 불러오기

```python
df = pd.read_csv('./data/2020_bitcoin_price.csv', encoding = 'cp949', names = ['Date', 'Price'])
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date').reset_index(drop = True)
```

```python
df
```

```python
train = df[:2066]
test = df[2066:]
```

# stationary한 시계열이 아니기 때문에 차분이 필요

```python
plt.plot(train['Date'], train['Price'])
```

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(data) :
    result = adfuller(data.values)
    print(f'ADF Statistics : {result[0]}')
    print(f'p-values : {result[1]}')
```

```python
adf_test(train['Price']) # 시계열이 안정적이지 않음
```

# 차분 

```python
diff_1 = train['Price'].diff().dropna()
plt.plot(diff_1)
plt.show()
```

```python
diff_2 = diff_1.diff().dropna()
plt.plot(diff_2)
plt.show()
```

```python
adf_test(diff_1)
```

```python
adf_test(diff_2) # 1회 차분, 2회 차분 모두 stationary 만족¶
```

## ACF, PACF 

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# 시각화
fig, ax = plt.subplots(1,2 , figsize = (12,4))

# acf plot
plot_acf(train['Price'], ax = ax[0]) # ACF는 점진적 감소 : p = 1

# pacf plot
plot_pacf(train['Price'], ax= ax[1]) # PACF는 자기자신을 제외한 첫번째 절단면 이후 절단면을 보임 : q = 0
plt.show()
```
